```python
# 加载图像并将其处理为224*224的大小
from PIL import Image
import os
ori_dataset_path = '/home/humw/Datasets/VGGFace2'
save_dataset_path = '/home/humw/Datasets/VGGFace2_224'
if not os.path.exists(save_dataset_path):
    os.makedirs(save_dataset_path)
for person_name in os.listdir(ori_dataset_path):
    person_path = os.path.join(ori_dataset_path, person_name, "set_B")
    save_person_path = os.path.join(save_dataset_path, person_name, "set_B")
    os.makedirs(os.path.join(save_dataset_path, person_name), exist_ok=True)
    os.makedirs(os.path.join(save_dataset_path, person_name, "set_B"), exist_ok=True)
    for img_name in os.listdir(person_path):
        img_path = os.path.join(person_path, img_name)
        save_img_path = os.path.join(save_person_path, img_name)
        img = Image.open(img_path)
        img = img.resize((224, 224))
        img.save(save_img_path)
```
